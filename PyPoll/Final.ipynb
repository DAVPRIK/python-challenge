```python
#import os module
import os
#import csv module
import csv

# Set path for file
csvpath = os.path.join("Resources", "election_data.csv")
#print(csvpath)

# Open and read the CSV
with open(csvpath, newline="") as csvfile:
        
    csvreader = csv.reader(csvfile, delimiter=",")
    csv_header = next(csvfile)
   
 
    #Declare variables as empty dictionaries and lists
    Candidates = {}
    Count = 0
    totalvotes = 0
    percent_of_votes = 0
    Most_Votes = 0
    Most_Voted = ""
    

    
    for row in csvreader:
        
        # Count the total number of votes cast
        candidate = row[2]
        Count += 1
        if candidate in Candidates.keys():
            Candidates[candidate] += 1
        else:
            Candidates[candidate] = 1
        #print(Candidates)
    
    
    # Print Statements
    print("Election Results")
    print("-------------------------------")
    print(f"Total Votes: {Count}")
    print("-------------------------------")
    
            
    #total number of votes for each candidate
    for candidate in Candidates:
        totalvotes += Candidates[candidate]
    
        # percent of votes for each candidate
        percent_of_votes = (Candidates[candidate])/(Count) * 100
        print(f"{candidate}: {int(percent_of_votes)}% {totalvotes}")
        
        if Candidates[candidate] > Most_Votes:
            Most_Voted = candidate
            Most_Votes = Candidates[candidate]
        
        
    
    # The winner of the election based on popular vote.
    print("-------------------------------")
    
    print(f"Winner: {Most_Voted}")
    
    print("-------------------------------")
    
    output_path = os.path.join("Resources", "output", "final.txt")

# Open the file using "write" mode. Specify the variable to hold the contents
with open(output_path, 'w', newline='') as final:

    # Initialize csv.writer

     #Write the first row (column headers)
    final.write("Election Results")
    final.write('\n')
    final.write("-------------------------------")
    final.write('\n')
    final.write(f"Total Votes: {Count}")
    final.write('\n')
    final.write("-------------------------------")
    final.write('\n')
    for candidate in Candidates:
        
        percent_of_votes = (Candidates[candidate])/(Count) * 100
        final.write(f"{candidate}: {int(percent_of_votes)}% {totalvotes}"'\n')
    final.write('\n')    
    final.write("-------------------------------")
    final.write('\n')
    final.write(f"Winner: {Most_Voted}")
    final.write('\n')
    #for candidate in Candidates:
        
     #   percent_of_votes = (Candidates[candidate])/(Count) * 100
      #  final.write(f"{candidate}: {int(percent_of_votes)}% {totalvotes}")
        
    final.write('\n')
    final.write("-------------------------------")
    

```
